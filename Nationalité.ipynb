```python
import sqlite3

# Connexion à la base de données SQLite
conn = sqlite3.connect('C:/Users/Cassandra/OneDrive/Documents/Université/UNINE/2024 - Printemps/Application de méthodes/Les_astronautes/astronautes.db')
cursor = conn.cursor()
```

```python
# Requête pour obtenir les nationalités
query = "SELECT nationalite FROM personne"
cursor.execute(query)

# Récupération des résultats
nationalites = cursor.fetchall()
```

```python
from collections import Counter

# Convertir la liste de tuples en une liste de nationalités
nationalites = [item[0] for item in nationalites]

# Compter les occurrences de chaque nationalité
compteur_nationalites = Counter(nationalites)

# Trouver la nationalité la plus représentée
nationalite_most_common = compteur_nationalites.most_common(1)
print("La nationalité la plus représentée est :", nationalite_most_common[0])
```

```python
import matplotlib.pyplot as plt

# Préparer les données pour le graphique
nationalites, counts = zip(*compteur_nationalites.items())

# Créer le graphique à barres
plt.figure(figsize=(10, 8))
plt.bar(nationalites, counts)
plt.xlabel('Nationalité')
plt.ylabel('Nombre d\'astronautes')
plt.title('Répartition des nationalités des astronautes')
plt.xticks(rotation=90)
plt.tight_layout()

# Afficher le graphique
plt.show()
```
