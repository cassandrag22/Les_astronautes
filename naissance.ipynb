```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connexion à la base de données SQLite
conn = sqlite3.connect('C:/Users/Cassandra/OneDrive/Documents/Université/UNINE/2024 - Printemps/Application de méthodes/Les_astronautes/astronautes.db')  

# Requête SQL pour extraire les dates de naissance
query = """
SELECT date_naissance 
FROM personne
"""

# Exécuter la requête et charger les résultats dans un DataFrame
df = pd.read_sql_query(query, conn)


```

```python
# Compter le nombre de naissances par année
birth_counts = df['year_of_birth'].value_counts().sort_index()
```

```python
# Création du graphique en barres
plt.figure(figsize=(12, 8))
plt.bar(birth_counts.index, birth_counts.values, color='skyblue')

# Titre et labels
plt.title('Distribution des naissances par année')
plt.xlabel('Année de naissance')
plt.ylabel('Nombre de naissances')

# Afficher le graphique
plt.show()
```
